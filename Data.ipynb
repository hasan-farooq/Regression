## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib as mtp
from matplotlib import pyplot as plt
import seaborn as sns
```

## Loading Data

```python
df = pd.read_csv('machine.data',header=None)
cols = ['vendor','model','myct','mmin','mmax','cach','chmin','chmax','prp','erp']
df.columns = cols
# df.head(2);
```

## Scatter Plot 
Tells how variables are scattered against the target variable

```python
scatter_plot = plt.figure(figsize=(10,5))
target = 'erp'
features = cols[2:9]
for feature in features:
    plt.scatter(x=df[feature],y=df[target],label=feature,marker='o')
plt.xlabel("Features")
plt.ylabel("CPU Relative Performance")
plt.legend()
# plt.xkcd()
plt.show()
```

## Correlation Matrix
Tells about the relation between variables

```python
plt.figure(figsize=(10,8))
sns.heatmap(data=df.corr(),annot=True)
plt.show()
```

## Pairs Plot
Tells about every variable.Like correlation matrix, it gives more insights to relation between variables

```python
sns.pairplot(df)
plt.show()
# open in new tab or download to zoom it
```

## Target Value
Gives idea on the value we will be predicting. In this case, it tells that there's some variable or variables that give a square shape. And also tells where data is concentrated and about outliers.

```python
plt.figure(figsize=(7,5))
indexes = []   # this is as a dummy x-axis
for i in range(len(df)):
    indexes.append(i)

plt.scatter(x=indexes,y=df[target],label=target,marker='o')
plt.xlabel("Index of Value")
plt.ylabel("Target Variable")
plt.legend()
plt.show()
```

## Comparing Errors

```python
models = ['Decision Tree','Pruned Decision Tree','Linear Regression','Polynomial Regression',
         'Random Forest','XGBoost Tree']
mse = [854,895,957,221,365,371]
rmse = [29,30,31,15,19,19]
mae = [15,15,21,6,12,10]
```
